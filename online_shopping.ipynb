```python
def add_item(item, price, quantity):
    if item in cart:
        # Use 'quantity' here to match the parameter name
        cart[item]["quantity"] += quantity
    else:
        # Use 'quantity' here as well
        cart[item] = {"price": price, "quantity": quantity}
```

```python
def update_quantity(cart,item,new_quantity):
  if item in cart:
    cart[item ]["quantity"]=new_quantity
  else:
    print("item not found")
```

```python
def remove_item (cart,item):
  if item in cart:
    del cart[item]
  else:
   print("item not found")
```

```python
cart = {}
add_item("Laptop",50000,1)
add_item("Mouse",500,3)
add_item("Keybord",1500,2)
print(cart)
```

```python

update_quantity(cart, "Keybord", 3)
print(cart)
```

```python
remove_item(cart, "Mouse")
print(cart)
```
